```python
import tkinter as tk
from pathlib import Path
from datetime import datetime
import os
import re
import pandas as pd
import numpy as np
from pandastable import Table

list_of_sheets = []

'''Colecting all xlsx files'''
Direc = os.getcwd()
#print(f"Files in the directory: {Direc}")
files = os.listdir(Direc)
files = [f for f in files if os.path.isfile(Direc+'/'+f)] #Filtering only the files.
r = re.compile(".*xlsx")
file_options = list(filter(r.match, files)) 

'''Creating the main window'''
window = tk.Tk()
window.geometry("750x250")
window.title('Welcome to VinkoExcel')

'''Buttons'''
button_start = tk.Button(text="Start!",#width=10,#height=5,
    bg="blue",fg="yellow", command = lambda: OnButtonClick(1)) 

button_select_sheet = tk.Button(text = 'Select', command = lambda: OnButtonClick(2))

button_end = tk.Button(text = 'End', command = lambda: OnButtonClick(3))

button_select_row = tk.Button(text = 'Select', command = lambda: OnButtonClick(4))


'''Option Box to chose which file you want and label'''
option_box = tk.Listbox(window)
file_select_label = tk.Label(text = 'Chose file.')
row_box = tk.Listbox(window)
sheet_select_label = tk.Label(text = 'Chose sheet.')


def Calloptions():
    option_box.pack(pady = 15)
    file_select_label.pack(side = 'top')
    button_select_sheet.pack(side = 'top')

    for file in range(len(file_options)):
        option_box.insert(file, file_options[file])

def Selected_item():
    for i in option_box.curselection():
        global file_name
        file_name = option_box.get(i)
        work_file = pd.ExcelFile(option_box.get(i))
        list_of_sheets = work_file.sheet_names
    return list_of_sheets

def Select_row():
    button_select_row.pack_forget()
    for option in row_box.curselection():
        #print(row_box.get(option))
        global excel_data_df
        excel_data_df = pd.read_excel(file_name, sheet_name = row_box.get(option))
        #display(excel_data_df)

def OnButtonClick( button_id):
    if button_id == 1:
        Calloptions()
        button_start.pack_forget()
        
    elif button_id == 2:
        list_of_sheets = Selected_item()
        button_select_sheet.pack_forget() 
        option_box.pack_forget()
        file_select_label.pack_forget()
        sheet_select_label.pack(side = 'top')
        row_box.pack(pady = 15)
        
        for sheet in range(len(list_of_sheets)):
            row_box.insert(sheet, list_of_sheets[sheet])
        button_select_row.pack(side = 'top')
            
    elif button_id == 3:
        window.destroy()
        
    elif button_id == 4:
        Select_row()

        
button_start.pack(side="top")




window.mainloop()
```
